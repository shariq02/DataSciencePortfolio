```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
# load car.csv dataset

df = pd.read_csv("Car_sales.csv")

df.head()
```

```python
sns.boxplot(x='Vehicle_type', y = 'Horsepower', data = df)
```

```python
sns.pairplot(df)
```

```python
# drop irrelevant columns

df = df.drop(['year_resale_value', 'Engine_size', 'Wheelbase', 'Width', 'Length', 'Curb_weight'], axis =1)

df.head()
```

```python
# Total number of rows and columns

df.shape
```

```python
# Rows containing duplicate data

duplicate_rows = df[df.duplicated()]
print('Number of duplicate rows:', duplicate_rows.shape)
```

```python
# drop duplicates

df = df.drop_duplicates()
df.count()
```

```python
# find null values

print(df.isnull().sum())
```

```python
# dropping the missing value

df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
# Findinf the outliners

sns.boxplot(x=df['Price_in_thousands'])
```

```python
sns.boxplot(x=df['Horsepower'])
```

```python
# Plotting a Histogram for number of cars per brand

df.Manufacturer.value_counts().nlargest().plot(kind='bar', figsize=(10,5))
plt.title('Number of cars by manufacture')
plt.xlabel('Manufacture')
plt.ylabel('Number of cars')
```

```python
# Find the relations between the variables

plt.figure(figsize=(20,10))
data = df.corr()
sns.heatmap(data,cmap='BrBG', annot=True)
data
```
