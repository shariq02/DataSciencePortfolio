```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset
```

```python
df = pd.read_csv("EPL_20_21.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# Create 2 new column
```

```python
df['MinsPerMatch'] = (df['Mins'] / df['Matches']).astype(int)
df['GoalsPerMatch'] = (df['Goals'] / df['Matches']).astype(float)
```

```python
df.head()
```

```python
# Total Goals

Total_Goals = df['Goals'].sum()
Total_Goals
```

```python
# Penalty Goals

Total_Penalty_Goals = df['Penalty_Goals'].sum()
Total_Penalty_Goals
```

```python
# Penalty Attempted

Total_Penalty_Attempted = df['Penalty_Attempted'].sum()
Total_Penalty_Attempted
```

```python
# Pie Chart for penalties missed vs scored

plt.figure(figsize=(13,6))
pl_not_scored = df['Penalty_Attempted'].sum() - Total_Penalty_Goals
data = [pl_not_scored, Total_Penalty_Goals]
lables = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = lables, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Unique Postion

df['Position'].unique()
```

```python
# Total number of FW players

df[df['Position'] == 'FW']
```

```python
# Total number of DF players

df[df['Position'] == 'DF']
```

```python
# Players from different nations

np.size((df['Nationality'].unique()))
```

```python
# Most players from which countries

nationality = df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette('magma'))
```

```python
# Clubs with maximum players in their squad

df['Club'].value_counts().nlargest(5).plot(kind = 'bar', color = sns.color_palette('viridis'))
```

```python
# Clubs with least players in their squad

df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color = sns.color_palette('viridis'))
```

```python
# Players based on age groups

under20 = df[df['Age'] <= 20]
age20_25 = df[(df['Age'] > 20) & (df['Age'] <= 25)]
age25_30 = df[(df['Age'] > 25) & (df['Age'] <= 30)]
above30 = df[df['Age'] > 30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
mylabels = ['<=20', '>20 & <=25', '>25 & <=30', '>30']
plt.title('Total Players with Age Groups', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = '%.1f%%')
plt.show()
```

```python
# Total number of player under 20 in each club

Players_under20 = df[df['Age'] <20]
Players_under20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette('cubehelix'))
```

```python
# Under 20 players in ManU

Players_under20[Players_under20['Club'] == 'Manchester United']
```

```python
# Under 20 players in Chelsea

Players_under20[Players_under20['Club'] == 'Chelsea']
```

```python
# Average age of players in each club

plt.figure(figsize = (12,6))
sns.boxplot(x = 'Club', y = 'Age', data = df)
plt.xticks(rotation = 90)
```

```python
num_player = df.groupby('Club').size()
data = (df.groupby('Club')['Age'].sum()) / num_player
data.sort_values(ascending = False)
```

```python
#  Total assist from each club

assist_by_club = pd.DataFrame(df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x = 'Club', y = 'Assists', data = assist_by_club.sort_values(by='Assists'), palette = 'Set2')
ax.set_xlabel('Club', fontsize=20)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs Vs Total Assists', fontsize = 20)
```

```python
# Top 10 Assists

top10_assists = df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n = 10, columns = 'Assists')
top10_assists
```

```python
# Total Goals from each Club

goals_by_clubs = pd.DataFrame(df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x = 'Club', y = 'Goals', data = goals_by_clubs.sort_values(by='Goals'), palette = 'Set2')
ax.set_xlabel('Club', fontsize=20)
ax.set_ylabel('Goals', fontsize=20)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs Vs Total Assists', fontsize = 20)
```

```python
# Most goals by player

top10_goals = df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n = 10, columns = 'Goals')
top10_goals
```

```python
#  Goal per match

top10_goals_per_match = df[['Name', 'GoalsPerMatch', 'Goals', 'Matches']].nlargest(n = 10, columns = 'GoalsPerMatch')
top10_goals_per_match
```

```python
# Pie Chart - Goals with assist and without assists

plt.figure(figsize=(14,7))
assists = df['Assists'].sum()
data = [Total_Goals - assists, assists]
lables = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(data, labels = lables, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards

epl_yellow_card = df.sort_values(by = 'Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most Yellow Cards')
c = sns.barplot( x = epl_yellow_card['Name'], y = epl_yellow_card['Yellow_Cards'], label = 'Players', color = 'yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
c
```

```python
# Top 10 players with most red cards

epl_red_card = df.sort_values(by = 'Red_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title('Players with most Red Cards')
c = sns.barplot( x = epl_red_card['Name'], y = epl_red_card['Red_Cards'], label = 'Players', color = 'red')
plt.ylabel('Number of Red Cards')
c.set_xticklabels(c.get_xticklabels(), rotation = 45)
c
```
